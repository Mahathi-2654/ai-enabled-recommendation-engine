```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("cleaned_reviews.csv")
df.head()
```

```python
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='product_id',
    values='rating'
)

user_item_matrix.head()
```

```python
user_item_matrix_filled = user_item_matrix.fillna(0)
user_item_matrix_filled.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

item_similarity = cosine_similarity(user_item_matrix_filled.T)

item_similarity_df = pd.DataFrame(
    item_similarity,
    index=user_item_matrix_filled.columns,
    columns=user_item_matrix_filled.columns
)

item_similarity_df.head()
```

```python
def recommend_products(product_id, n=5):
    """
    Recommends top-N similar products for a given product_id
    using item-item cosine similarity.
    """
    # Check if product exists
    if product_id not in item_similarity_df.columns:
        return "Product not found"

    # Get similarity scores for the given product
    similarity_scores = item_similarity_df[product_id].sort_values(ascending=False)

    # Exclude the product itself and get top-N recommendations
    recommendations = similarity_scores.iloc[1:n+1]

    return recommendations
```

```python
recommend_products("7310172001", n=5)
```
