```python
import pandas as pd
```

```python
data = pd.read_csv(
    "Reviews.csv",
    engine="python",
    on_bad_lines="skip"
)
```

```python
data.rename(columns={
    'UserId': 'user_id',
    'ProductId': 'product_id',
    'Score': 'rating'
}, inplace=True)
```

```python
data = data[['user_id', 'product_id', 'rating']]
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data = data.sample(30000, random_state=42)
```

```python
user_item_matrix = data.pivot_table(
    index='user_id',
    columns='product_id',
    values='rating',
    fill_value=0
)
```

```python
print("✅ Milestone 1 completed successfully")
print(user_item_matrix.head())
print("Matrix shape:", user_item_matrix.shape)
```

```python
# Save cleaned dataset
data.to_csv("cleaned_reviews.csv", index=False)

# Save user-item interaction matrix
user_item_matrix.to_csv("user_item_matrix.csv")

print("Files saved successfully")
```
